## 인기 영화 추천 시스템

https://www.kaggle.com/datasets/tmdb/tmdb-movie-metadata

Dataset
영화에 대한 정보(예산, 장르, 평점, 런닝시간, 제목, 키워드 등)의 영화 연출팀과 배우들에 대한 별도 정보로 구성

- 인기 영화 추천 알고리즘 작성 인기 컨텐츠와 동일
- 장르를 기준으로 인기영화를 추천


```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('https://grepp-reco-test.s3.ap-northeast-2.amazonaws.com/tmdb_5000_movies.csv')
movies.head()
```

```python
movies.shape
```

```python
movies.describe()
```

```python
credits = pd.read_csv("https://grepp-reco-test.s3.ap-northeast-2.amazonaws.com/tmdb_5000_credits.csv")
credits.head()
```

```python
import json


def add_genre_name(j):
    genres = []
    arr = json.loads(j)
    for a in arr:
        genres.append(a.get('name'))

    return " ".join(sorted(genres))

movies['genre_names'] = movies.apply(lambda x: add_genre_name(x.genres), axis=1)
```

```python
movies['genre_names']
```

```python
movies['genre_names'].nunique()
```

```python
movies.groupby('genre_names').size()
```

```python
credits.head()
```

```python
credits.shape
```

```python
# movies credits inner join
movie_credits = pd.merge(movies, credits, left_on="id", right_on="movie_id")
movie_credits.head()
```

```python
movie_credits.shape
```

```python
movie_credits.drop(
    columns=[
        "homepage", "title_x", "title_y",
        "status", "production_countries", "production_companies"
    ],
    inplace=True
)
movie_credits.head()
```

```python
movie_credits.info()
```

```python
movie_credits.describe()
```

```python
popularity = movie_credits.sort_values('popularity', ascending=False)
popularity
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(
    x=popularity['popularity'].head(10),
    y=popularity['original_title'].head(10)
)

plt.title("Most Popular by Popularity", weight="bold")
plt.xlabel("Score of Popularity", weight="bold")
plt.ylabel("Movie Title", weight="bold")
plt.savefig("best_popular_movies.png")
```

```python
def reco_top_scored_one(n, genre=None):
    if genre is None:
        return popularity['original_title'].head(n)
    else:
        return popularity[popularity['genre_names'].str.contains(genre)]['original_title'].head(n)

reco_top_scored_one(10)
```

```python
reco_top_scored_one(10, "Science Fiction")
```
